```python
import pandas as pd, numpy as np
```

```python
# we want to get the final list of genes we want to connect to dgidb with
with open('./gene datasets/coronavirus-functional-enriched.csv') as f:
    df = pd.read_csv(f, sep=',')
```

```python
df
```

```python
unique_genes = df['name'].unique()
print(f'Number of unique gene names: {unique_genes.size}')
```

```python
# save unique genes list as a csv file
with open('./gene datasets/coronavirus-functional-enriched-genes.csv', 'w+') as f:
    np.savetxt(f, unique_genes, delimiter=',', fmt='%s')
    print('File written successfully!')
```
