```python
import pandas as pd, numpy as np
```

```python
with open('./gene datasets/coronavirus.tsv') as f:
    df = pd.read_csv(f, sep='\t')
```

```python
df
```

```python
unique_genes = df['Associated Gene Name'].unique()
print(f'Number of unique gene names: {unique_genes.size}')
```

```python
# save unique genes list as a csv file
with open('./gene datasets/coronavirus-unique.csv', 'w+') as f:
    np.savetxt(f, unique_genes, delimiter=',', fmt='%s')
    print('File written successfully!')
```
